```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the CSV file.
df=pd.read_csv("D:/Machine Learning Practical/Lab Exercises/Lab 1/Program1/college.csv")
df.head()
```

```python
#Setting index as Names of the Universities
college=df.set_index('Names')
college.head()
```

```python
#Checking for NULL VALUES
college.isnull().sum()

#dropping null values
#df2.dropna(subset=['Population'])
```

Interpretation: We see that there are no null values in the dataset

```python
#1. SUMMARY
college.describe()
```

Interpretation: We can see the summary of the dataset we have for each of the FEATURES.

```python
#2. SCATTER PLOT MATRIX of the first 10 columns

#Slicing the first 10 rows
first_ten=college.iloc[:,1:10]
#Plotting 
plot1=pd.plotting.scatter_matrix(first_ten)

#Saving Plot as Image
plt.savefig("D:/Machine Learning Practical/Lab Exercises/Lab 1/ScatterPlot.png")
```

Interpretation: From the scatter plot we observe that mist of the plots are irregular ,while some follow linearity.

```python
#3. Side by Side BOXPLOT 

#Creating 2 dataframes based on Private COlumn
box_11=college[college["Private"]=='Yes']
box_12=college[college["Private"]=="No"]

#Creating Boxplot
sns.boxplot(data=[box_11["Outstate"],box_12["Outstate"]],
    palette=[sns.xkcd_rgb["pale red"],sns.xkcd_rgb["medium green"]],
    showmeans=True,)
               
#Saving Figure
plt.savefig("D:/Machine Learning Practical/Lab Exercises/Lab 1/BoxPlot1.png")
```

Interpretation: The boxplot tell us about the interquartile range for the OUTSTATE feature based on if the college is Private or not. We can also see the outliers in the dataset.

```python
#4. ELITE

#Creating Elite
list1=[0]
Elite=list(np.repeat(list1,len(college["Top10perc"])))

for i in range(0,len(Elite)):
    if college["Top10perc"][i]>50:
        Elite[i]=1
        
Elite=pd.DataFrame(Elite)

#Rabsenaming column name
Elite.rename(columns={0:'Elite'},inplace=True)

#No. OF Elite Universities
Elite.value_counts()
```

Interpretation: There are 78 Elite schools

```python
#Boxplot for ELite vs Outstate

#Adding Elite column to College
l1=list(Elite["Elite"])
college["Elite"]=l1

#CReating 2 Dataframes based on ELite column
#Creating 2 dataframes based on Private COlumn
box_21=college[college["Elite"]==1]
box_22=college[college["Elite"]==0]

#Creating Boxplot
sns.boxplot(
    data=[box_21["Outstate"],box_22["Outstate"]],
    palette=[sns.xkcd_rgb["pale red"],sns.xkcd_rgb["medium green"]],
    showmeans=True
,)
               
#Saving Figure
plt.savefig("D:/Machine Learning Practical/Lab Exercises/Lab 1/BoxPlot2.png")
```

```python
#5. HISTOGRAM

plt.subplot(2,2,1)
plt.hist(college["PhD"])
plt.title("PhD")

plt.subplot(2,2,2)
plt.hist(college["Grad.Rate"])
plt.title("Grad.Rate")

plt.subplot(2,2,3)
plt.hist(college["Books"])
plt.title("Books")

plt.subplot(2,2,4)
plt.hist(college["Terminal"])
plt.title("Terminal")

plt.savefig("D:/Machine Learning Practical/Lab Exercises/Lab 1/Histogram.png")
```

Interpretation: We plotted the histogram for 4 different features, which tells us the frequency of each of the feature based on different classes. We can check normality using these graphs
