## DECISION TREES

### Decision Tree Classifier:

```python
#Importing necessary libraries:
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import export_text
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import graphviz
import pandas as pd

#Iris dataset:
iris = load_iris()
iris
```

```python
#Splitting into features and target variable:
X, y = iris.data, iris.target

#Train and Test Data
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=1)

#Iris has categorical target variable with target classes: setosa, versicolor, virginica.

#Fitting the decision tree classifier:
a = tree.DecisionTreeClassifier()
clf = a.fit(X_train, y_train)
```

```python
#plot the tree:
plt.figure(figsize=(10,6))
tree.plot_tree(clf)
```

```python
#Training and testing accuracy
print("The Training Accuracy is : ",clf.score(X_train, y_train))
print("The Testing Accuracy is : ",clf.score(X_test, y_test))
```

Note: We see that the training accuracy is 1 because the model is overfitted.Hence we need to find the optimal maximum depth of the tree

```python
#We wont be doing GRID_SEARCH_CV since we only have Max_Depth as a parameter

for i in range(1,10):
      
      clf = tree.DecisionTreeClassifier(max_depth=i).fit(X_train,y_train)
      print("Training score for max depth",i,"is: ",clf.score(X_train, y_train),"\n")
      print("Testing score for max depth",i,"is: ",clf.score(X_test, y_test),"\n")
```

Note: We see that for Max Depth =3 , we get the optimal accuracy without overfitting. Hence that is the best model.
    

```python
#Fitting the best model
a= tree.DecisionTreeClassifier(max_depth=3)
clf_best = a.fit(X_train, y_train)

#Training and testing accuracy
print("The Training Accuracy is : ",clf_best.score(X_train, y_train))
print("The Testing Accuracy is : ",clf_best.score(X_test, y_test))
```

```python
dot_data = tree.export_graphviz(clf_best, out_file=None, 
                     feature_names=iris.feature_names,  
                     class_names=iris.target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
#Simpler implementation:
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=3)
decision_tree = decision_tree.fit(iris.data, iris.target)
r = export_text(decision_tree, feature_names=iris['feature_names'])
print(r)
```

```python
#Prediction Using The best model
clf_best.predict(X_test)
```

### Decision Tree Regressor:

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor as DTR
import matplotlib.pyplot as plt
```

```python
#Car Dataset
car=pd.read_csv("D:/Machine Learning Practical/Lab Exercises/Decision Tree/car1.csv")
car
```

```python
#Train Test Data
X=pd.DataFrame(car.drop(["Sales"],axis=1))
y=pd.DataFrame(car["Sales"])

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
# Fit regression model with different depths

from sklearn.metrics import mean_squared_error

for i in range(1,11):
      reg = DTR(max_depth=i)
      dtr = reg.fit(X_train,y_train)
    
      y_pred=dtr.predict(X_test)
      print("Training score for max depth",i,"is: ",mean_squared_error(y_test,y_pred),"\n")
      print("Testing score for max depth",i,"is: ",mean_squared_error(y_test,y_pred),"\n")
    
```

Note: Therefore the best model is for Max Depth=2 since the MSE is least for it

```python
#Fitting the best model

reg = DTR(max_depth=2)
dtr_best = reg.fit(X_train,y_train)

#Predicting the test values
y_pred=dtr_best.predict(X_test)
```

```python
dot_data = tree.export_graphviz(dtr_best, out_file=None,  
                     class_names="Sales",  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
